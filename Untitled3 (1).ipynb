```python
data=data.copy()
data['high_humidity_level']=(data['relative_humidity_3pm']>24.99)*1
data['high_humidity_level'].head()
```

```python
y=data[['high_humidity_level']].copy()
y.head()
```

```python
morning_feature=['air_pressure_9am', 'air_temp_9am', 'avg_wind_speed_9am',
       'avg_wind_direction_9am', 'max_wind_direction_9am',
       'max_wind_speed_9am', 'rain_accumulation_9am', 'rain_duration_9am',
       'relative_humidity_9am']
```

```python
x=data[morning_feature].copy()
x.columns
```

```python
y.columns
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=324)
```

```python
humidity_classifier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=1)
humidity_classifier.fit(x_train,y_train)
```

```python
type(humidity_classifier)
```

```python
y_predicted= humidity_classifier.predict(x_test)
```

```python
y_test['high_humidity_level'][:10]
```

```python
accuracy_score(y_test,y_predicted)*100
```
